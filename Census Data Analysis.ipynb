## Employment and Unemployment

```python
# This functions categorizes the various occupation into groups for ease of analysis
def employment_grouping(row):
    if row['Occupation'] == 'Child':
        return 'Not Applicable'
    elif row['Occupation'] == 'Student':
        return 'Student'
    elif row['Occupation'] == 'University Student':
        return 'Student'
    elif row['Occupation'] == 'PhD Student':
        return 'Student'
    elif 'Retired' in row['Occupation']:
        return 'Retired'
    elif row['Occupation'] == 'Unemployed' and row['Age'] > 65:
        return 'Retired'
    elif row['Occupation'] == 'Unemployed':
        return 'Unemployed'
    else:
        return 'Employed'

df['Grouped Occupation'] = df.apply(lambda row: employment_grouping(row), axis=1)
```

## Death Rate

```python
#Old age age interval
age_category = []

skip = 5

for i in range (56,105,skip):
    end = i + skip-1
    age_category.append (f'{i} - {end}')
    
print(age_category)
```

```python
#Number of old people within the selected age band
num_old_age = []

for i in age_category:
    start, end = i.split('-')
    age_range_df = df[(df['Age'] >= int(start)) & (df['Age'] <= int(end))]
    num_old_age.append(age_range_df['Age'].count())

print(num_old_age)
```

```python
#Number of deaths within the selected age bands

num_of_deaths = []

for i in range(0, len(num_old_age)-1,2):
    output = num_old_age[i] - num_old_age[i+1]
    num_of_deaths.append(output)

print(num_of_deaths)
```

## Occupancy

```python
# get the number of unique houses based on street and house number
num_houses = df.groupby(['Street', 'House Number']).size().reset_index(name='Count')['House Number'].count()

print(num_houses)
```

```python
# This groups the unique House Number and street and First Name to obtain the Number of houses
typical_occupancy = df.groupby(["House Number", "Street"])["First Name"].nunique()

# Create a new dataframe for the number of houses and renames firstname to Typical count
df_typical_occupancy = pd.DataFrame(typical_occupancy).reset_index().rename(columns={"First Name": "Typical Count"})

# creating the dataframe
df_typical_occupancy = df_typical_occupancy[["House Number", "Typical Count"]]

# Calculate the average count
average_occupancy = round(typical_occupancy.mean())

# This adds a new column "Average Count" and assign it the value of average_count in the dataframe
df_typical_occupancy["Average Count"] = average_occupancy

df_typical_occupancy
```

```python
# number of overcrowded houses

crowded_houses = df_typical_occupancy[df_typical_occupancy["Typical Count"] > df_typical_occupancy["Average Count"]]['Average Count'].count()

crowded_houses
```

```python
#percentage of overcrowded houses

percentage_crowded_houses = (crowded_houses / num_houses) * 100

percentage_crowded_houses
```

```python
#number of underoccupied houses

under_occupied = df_typical_occupancy[df_typical_occupancy["Typical Count"] < df_typical_occupancy["Average Count"]]['Average Count'].count()

under_occupied
```

```python
#percentage of underoccupied houses

percentage_underoccupied_houses = (under_occupied / num_houses) * 100

percentage_underoccupied_houses
```

```python
# Number of perfectly ocuppied houses

exact_occupancy = df_typical_occupancy[df_typical_occupancy["Typical Count"] == df_typical_occupancy["Average Count"]]['Average Count'].count()

exact_occupancy
```

```python
# percenge of perfectly ocuppied houses

percentage_exact_occupancy = (exact_occupancy / num_houses) * 100

percentage_exact_occupancy
```
